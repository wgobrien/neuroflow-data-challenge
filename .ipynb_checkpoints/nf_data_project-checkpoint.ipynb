```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read in csv
gad = pd.read_csv("phq_all_final.csv")

# convert date columns to datetime objects
gad.date = pd.to_datetime(gad.date)
gad.patient_date_created = pd.to_datetime(gad.patient_date_created)

gad.head()
```

To get an idea of the code we're working with, I've made a distribution of the number of tests that each patient in the system has taken. This gives us a sense of the data points per patient and what we should be aiming to derive from the data set.

```python
plt.figure(figsize=(10,6))

patient_counts = gad.groupby(by=["patient_id"])["patient_id"].count()
mean_num = np.mean(patient_counts)
median_num = np.median(patient_counts)

n = plt.hist(patient_counts, bins=50, density=True)
plt.vlines(mean_num, 0, max(n[0]), color="red", lw=2, label=f"mean: {mean_num}")
plt.vlines(median_num, 0, max(n[0]), color="magenta", lw=2, label=f"median: {median_num}")

plt.title("Distribution of Tests Taken", fontsize=16)
plt.ylabel("Frequency", fontsize=14)
plt.xlabel("Number of Assesments Taken", fontsize=14)

plt.grid()
plt.legend(fontsize=14)
plt.show()
```

When analyzing patient progress, one issue that arises with this data is a lack of scores for certain patients. The average patient has taken about 3.5 tests. The median is 3.0. This low number of data points per patient may be a hard metric to use for progress on an individual basis. Even when plotting the patients with many scores, the scores can vary greatly and it can be hard to derive insights from that alone. Let's plot the data for the patients with the top five most number of scores as an example: 

```python
# extract the 5 patient ids with highest number of assesments taken
ids = gad.groupby(by=["patient_id"])["patient_id"].count().sort_values(ascending=False).head(5)

plt.figure(figsize=(10,6))

for pat_id in ids.index:
    patient_id = gad[gad.patient_id == pat_id]
    plt.plot(patient_id.date, patient_id.score, label=f"{pat_id}")

plt.title("Selected Patient Scores Over Time", fontsize=16)
plt.ylabel("Score", fontsize=14)
plt.xlabel("Date of Test", fontsize=14)

plt.grid()
plt.legend(title="Patient IDs")
plt.show()
```

Based on these graphs, it is difficult to find a pattern of progress over time. Some have increases in the score, some fluctuate between a low range, and others decrease and increase sporadically. Also, since it will also be hard to derive a pattern for the patients with lower numbers of tests, it will be best to create a new metric based on the duration of time that a patient has been in the program. To measure progress, I'll look at the net change between the earliest test and the most recent test taken by each patient. I won't look at patients with only one test since we can't get a reasonable indication of progress from one data point. By looking at duration rather than number of tests, we can show that if a patient has taken 3 tests 40 days apart, we can still assume that they have been working on exercises in between, so we can eliminate the problem of low number of tests per person.

```python
# retrieve duration of time in program
time_in_program = (gad.groupby(by=["patient_id"])["date"].max()\
                   - gad.groupby(by=["patient_id"])["patient_date_created"].min()).dt.days

# scores, dates, ids for first test taken 
first_date = gad.loc[gad.groupby("patient_id")["date"].idxmin()][["patient_id","date", "score"]].set_index("patient_id")

# scores, dates, ids for most recent test taken
recent_date = gad.loc[gad.groupby("patient_id")["date"].idxmax()][["patient_id","date", "score"]].set_index("patient_id")

# retrieve net change for each patient
score_deltas = recent_date.score-first_date.score

# concatenate duration of time with score changes
time_delta = pd.DataFrame()
time_delta["duration"] = time_in_program
time_delta["score_change"] = score_deltas
```

```python
pos = time_delta[time_delta.score_change >= 0]
neg = time_delta[time_delta.score_change < 0]

plt.figure(figsize=(10,6))

plt.scatter(pos.duration, pos.score_change, alpha=.6, c="r")
plt.scatter(neg.duration, neg.score_change, alpha=.6, c="b")

plt.title("Juxtaposed")
plt.ylabel("Change in Score", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)
plt.grid()
plt.show()
```

```python
change_over_time = time_delta.groupby(by=["duration"]).min().reset_index()

x = change_over_time.duration
y = change_over_time.score_change
z = np.polyfit(x, y, 2)

f = np.poly1d(z)

plt.figure(figsize=(10,6))

plt.scatter(x, y, alpha=.6)
plt.plot(x, f(x), c="r")

plt.ylabel("Change in Score", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)
plt.grid()
plt.show()
```

```python
change_over_time = time_delta.groupby(by=["duration"]).max().reset_index()

x = change_over_time.duration
y = change_over_time.score_change
z = np.polyfit(x, y, 2)

f = np.poly1d(z)

plt.figure(figsize=(10,6))

plt.scatter(x, y, alpha=.6)
plt.plot(x, f(x), c="r")

plt.ylabel("Change in Score", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)
plt.grid()
plt.show()
```

```python
change_over_time = time_delta.groupby(by=["duration"]).mean().reset_index()

x = change_over_time.duration
y = change_over_time.score_change
z = np.polyfit(x, y, 1)

f = np.poly1d(z)

plt.figure(figsize=(10,6))

plt.scatter(x, y, alpha=.6)
plt.plot(x, f(x), c="r")

plt.title("Average Changes Over Time")
plt.ylabel("Change in Score", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)
plt.grid()
plt.show()
```
