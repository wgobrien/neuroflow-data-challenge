```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read in csv
gad = pd.read_csv("part1/phq_all_final.csv")

# convert date columns to datetime objects
gad.date = pd.to_datetime(gad.date)
gad.patient_date_created = pd.to_datetime(gad.patient_date_created)

gad.head()
```

To get an idea of the code we're working with, I've made a distribution of the number of tests that each patient in the system has taken. This gives us a sense of the data points per patient and what we should be aiming to derive from the data set.

```python
plt.figure(figsize=(10,6))

# retrieve number of tests taken per patient
patient_counts = gad.groupby(by=["patient_id"])["patient_id"].count()
mean_num = np.mean(patient_counts)
median_num = np.median(patient_counts)

n = plt.hist(patient_counts, bins=50, density=True)
plt.vlines(mean_num, 0, max(n[0]), color="red", lw=2, label=f"mean:    {round(mean_num, 3)}")
plt.vlines(median_num, 0, max(n[0]), color="magenta", lw=2, label=f"median: {median_num}")

plt.title("[I] Distribution of Tests Taken", fontsize=16)
plt.ylabel("Frequency", fontsize=14)
plt.xlabel("Number of Assesments Taken", fontsize=14)

plt.grid()
plt.legend(fontsize=14)
plt.show()
```

When analyzing patient progress, one issue that arises with this data is a lack of scores for certain patients. The average patient has taken about 3.5 tests. The median is 3.0. This low number of data points per patient may be a hard metric to use for progress on an individual basis. Even when plotting the patients with many scores, the scores can vary greatly and it can be hard to derive insights from that alone. Let's plot the data for the patients with the top five most number of scores as an example: 

```python
# extract the 5 patient ids with highest number of assesments taken
ids = gad.groupby(by=["patient_id"])["patient_id"].count().sort_values(ascending=False).head(5)

plt.figure(figsize=(10,6))

# plot the individual patients scores over time
for pat_id in ids.index:
    patient_id = gad[gad.patient_id == pat_id]
    plt.plot(patient_id.date, patient_id.score, label=f"{pat_id}")

plt.title("[II] Selected Patient Scores Over Time", fontsize=16)
plt.ylabel("Score", fontsize=14)
plt.xlabel("Date of Test", fontsize=14)

plt.grid()
plt.legend(title="Patient IDs")
plt.show()
```

Based on these graphs, it is difficult to find a pattern of progress over time. Some have increases in the score, some fluctuate over a small range of scores, and others decrease and increase sporadically. Also, since it will  be difficult to derive a pattern for the patients with low numbers of tests (which represents the majority of patients), it will be best to look at the data in a more wholistic way. I'm going to create a new metric based on the duration of time that a patient has been in the program. To measure progress, I'll look at the net change between the earliest test and the most recent test taken by each patient. I won't look at patients with only one test since we can't get a reasonable indication of progress from one data point. By looking at duration rather than total number of tests taken, we can show that if a patient has taken only 3 tests 40 days apart versus 20 tests 6 days apart, for example, we can assume that both patients have been working on exercises with clinicians in between tests to improve their mental health. This eliminates the problem of low number of tests per person, allows us to compare between patients with differing numbers of tests, and gives us a better measure of how patients are progressing.

```python
# dropping patients with only one score, can't measure progress off of a single test and
# that data would bloat the visualizations
gad_edit = gad[~gad['patient_id'].isin(patient_counts[patient_counts < 2].index)]

# retrieve duration of time in program for each patient
time_in_program = (gad_edit.groupby(by=["patient_id"])["date"].max()\
                   - gad_edit.groupby(by=["patient_id"])["patient_date_created"].min()).dt.days

# scores, dates, ids for first tests taken 
first_date = gad_edit.loc[gad_edit.groupby("patient_id")["date"].idxmin()][["patient_id","date", "score"]].set_index("patient_id")

# scores, dates, ids for most recent tests taken
recent_date = gad_edit.loc[gad_edit.groupby("patient_id")["date"].idxmax()][["patient_id","date", "score"]].set_index("patient_id")

# retrieve change in score for each patient from beginning to present
score_deltas = recent_date.score-first_date.score

# concatenate duration of time with score changes into new dataframe
time_delta = pd.DataFrame()
time_delta["duration"] = time_in_program
time_delta["score_change"] = score_deltas
```

Now that we have a new dataframe with the score changes and duration of time in the program for each patient, we can plot this data and see how the change in score is related to the duration of time in the program. I'm going to further aggregate the patients based on duration of time such that patients in the program for the same number of days will have their progress measured together. I've done three aggregations of the maximums, minimums, and means of the score changes per duration of time. This reduces the number of data points and gives a more clear visualization of the data.

**Minimum aggregate of score changes for duration of time**

```python
# aggregate by largest score decreases for a duration of time
change_over_time = time_delta.groupby(by=["duration"]).min().reset_index()

x = change_over_time.duration
y = change_over_time.score_change
z_2 = np.polyfit(x, y, 2)
z_3 = np.polyfit(x, y, 3)

f = np.poly1d(z_2)
g = np.poly1d(z_3)

plt.figure(figsize=(10,6))

plt.scatter(x, y, alpha=.6, label=f"$r^2$: {round(np.corrcoef(x,y)[0,1], 5)}")
# fit polynomial with degree 2
plt.plot(x, f(x), c="r", label="deg 2")
# fit polynomial with degree 3, may be subject to overfitting when we find more data
plt.plot(x, g(x), c="c", label="deg 3")

plt.title("[III] Largest Decreases In Score Over Time",fontsize=16)
plt.ylabel("Change in Score", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)

plt.legend()
plt.grid()
plt.show()
```

When aggregating by the minimum values, we can see that we've extracted the largest improvements in score for patients per duration of time. To make this more clear, one data point represents the largest decrease that all patients had in the program for a certain number of days in the program.

How is the useful?

The correlation coefficient for this metric is $r^2 \approx -0.60$, which means that patients who are showing larger improvement tend to continue to improve the longer they stay in the program and stick with the exercies over extended periods of time.

**Maximum aggregate of score changes for duration of time**

```python
# aggregate by largest score increases for a duration of time
change_over_time = time_delta.groupby(by=["duration"]).max().reset_index()

x = change_over_time.duration
y = change_over_time.score_change
z_2 = np.polyfit(x, y, 2)
z_3 = np.polyfit(x, y, 3)

f = np.poly1d(z_2)
g = np.poly1d(z_3)

plt.figure(figsize=(10,6))

plt.scatter(x, y, alpha=.6, label=f"$r^2$: {round(np.corrcoef(x,y)[0,1], 5)}")
# fit polynomial with degree 2
plt.plot(x, f(x), c="r", label="deg 2")
# fit polynomial with degree 3, may be subject to overfitting when we find more data
plt.plot(x, g(x), c="c", label="deg 3")

plt.title("[IV] Largest Increases In Score Over Time",fontsize=16)
plt.ylabel("Change in Score", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)

plt.legend()
plt.grid()
plt.show()
```

When aggregating by the maximum values, we've extracted the largest increases in score for patients per duration of time in the program.

How is the useful?

The correlation coefficient for this metric is $r^2 \approx 0.47$, which means that for those patients who achieved worse scores from start to finish, the scores tend to get worse over extended periods of time. This is a weaker correlation than in the previous metric which had the opposite effect.

Between charts [III] and [IV], it seems that there is an effect of momementum that occurs where patients whose scores changed to a larger degree in a certain direction tended to continue to move in the same direction. In other words, patients who showed large improvements tended to continue to improve, while patients who worsened to a larger degree would also tend to worsen over time. However, this metric is not perfectâ€“ there are issues with it that I will explain later on.

**Mean aggregate of score changes for duration of time**

```python
change_over_time = time_delta.groupby(by=["duration"]).mean().reset_index()

x = change_over_time.duration
y = change_over_time.score_change
z = np.polyfit(x, y, 1)

f = np.poly1d(z)

plt.figure(figsize=(10,6))

plt.scatter(x, y, alpha=.6, label=f"$r^2$: {round(np.corrcoef(x,y)[0,1], 5)}")
plt.plot(x, f(x), c="r", label=f"m: {round(f[1],5)}")

plt.title("[V] Average Score Change Over Time",fontsize=16)
plt.ylabel("Change in Score", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)

plt.legend()
plt.grid()
plt.show()
```

This chart takes a look at the average change in score for a duration in time. There is a small negative correlation of $r^2 \approx -0.22$ which suggests that patients are generally tending to improve between their first and most recent tests the longer they stay in the program. However, the rate at which the scores tend to decrease is very small as shown by the slope of the best fit, and this is where this approach starts to fall apart.

While this approach provides some useful insights, we run into issues where we aren't representing patients with large fluctuations between the first and last tests. For example, a patient who has scores of 5, 15, and 6 is shown to have increased in score by a single point. However, there is the obvious issue of the spike to 15 in the middle, which would require further evaluation from a clinician.

One way we can handle this is by plotting the means of the scores of each patient over time. This will give a more robust and wholistic review of patient performance by considering every test taken. To do this, I will first group the patients by ID and find the mean scores. I will then concatenate this with the duration of time as indexed by the IDs. Next, I will further aggregate the resulting dataframe by duration of time and take the mean value so that the data weighs the outliers more heavily.

This is different than the approach above because instead of looking at just the start and finish, we will be averaging each patient's scores and analyzing that with respect to the duration of time in the program.

**Below we take the mean, minimum, and maximums for this new strategy and visualize the results:**

```python
# duration of time in program
time_in_program = ((gad_edit.groupby(by=["patient_id"])["date"].max()\
                   - gad_edit.groupby(by=["patient_id"])["patient_date_created"].min()).dt.days).reset_index()

time_in_program.columns = ["patient_id","duration"]

patient_scores = pd.DataFrame()
patient_scores["duration"] = time_in_program.duration

# means of each patient's scores
patient_mean_scores = gad_edit.groupby("patient_id").mean().reset_index()
patient_scores["means"] = patient_mean_scores.score

# standard deviations of each patient's scores for use later on
patient_mean_scores = gad_edit.groupby("patient_id").std().reset_index()
patient_scores["stdev"] = patient_mean_scores.score

# group by duration of time, take mean value to weight outliers more heavily
grouped = patient_scores.groupby(by=["duration"]).mean().reset_index()

x = grouped.duration
y = grouped.means
z = np.polyfit(x, y, 2)

f = np.poly1d(z)

plt.figure(figsize=(10,6))

plt.scatter(x, y, alpha=.6, label=f"$r^2$: {round(np.corrcoef(x,y)[0,1], 5)}")
plt.plot(x, f(x), c="r")

plt.title("[VI] Averages of Patient Scores For Duration of Time in Program",fontsize=16)
plt.ylabel("Averages of Patient Scores", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)

plt.legend()
plt.grid()
plt.show()
```

```python
# means of each patient's scores
patient_mean_scores = gad_edit.groupby("patient_id").min().reset_index()
patient_scores["means"] = patient_mean_scores.score

# group by duration of time, take mean value to weight outliers more heavily
grouped = patient_scores.groupby(by=["duration"]).mean().reset_index()

x = grouped.duration
y = grouped.means
z = np.polyfit(x, y, 2)

f = np.poly1d(z)

plt.figure(figsize=(10,6))

plt.scatter(x, y, alpha=.6, label=f"$r^2$: {round(np.corrcoef(x,y)[0,1], 5)}")
plt.plot(x, f(x), c="r")

plt.title("[VII] Minimums of Patient Scores For Duration of Time in Program",fontsize=16)
plt.ylabel("Averages of Patient Scores", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)

plt.legend()
plt.grid()
plt.show()
```

```python
# means of each patient's scores
patient_mean_scores = gad_edit.groupby("patient_id").max().reset_index()
patient_scores["means"] = patient_mean_scores.score

# group by duration of time, take mean value to weight outliers more heavily
grouped = patient_scores.groupby(by=["duration"]).mean().reset_index()

x = grouped.duration
y = grouped.means
z = np.polyfit(x, y, 2)

f = np.poly1d(z)

plt.figure(figsize=(10,6))

plt.scatter(x, y, alpha=.6, label=f"$r^2$: {round(np.corrcoef(x,y)[0,1], 5)}")
plt.plot(x, f(x), c="r")

plt.title("[VIII] Maximums of Patient Scores For Duration of Time in Program",fontsize=16)
plt.ylabel("Maximums of Patient Scores Per ", fontsize=14)
plt.xlabel("Days in Program", fontsize=14)

plt.legend()
plt.grid()
plt.show()
```

This data reveals some exciting insights. For the mean sores, there is a correlation of $r^2 \approx -0.65$ which indicates that the program and exercises that NeuroFlow is providing is revealing improvement in patient scores over time. The charts with the maximums (low negative correlation -0.32) and minimums (high negative correlation -0.77) tell us that those who have a lower score range have a tendancy to improve at a faster rate than those in a higher score range as shown by the correlation values of each respective chart. Both the maximum and the minimum have negative correlations, which means patients are improving the longer they stay in the program.

```python
dev = gad_edit.groupby("patient_id").std().reset_index()

grouped = patient_scores.groupby(by=["duration"]).median().reset_index()

plt.figure(figsize=(10,6))

plt.hist(patient_scores.stdev)

plt.title("[VII] Standard Deviaitions of Patient Scores",fontsize=16)
plt.ylabel("Frequency", fontsize=14)
plt.xlabel("Standard Deviation", fontsize=14)

plt.grid()
plt.show()
```

This distribution shows the standard deviations of patient scores, ultimately showing that the majority of patients are not rapidly shifting between different levels of severity. This provides some confidence to our last results in that the mean values are not being heavily influenced by large fluctuations in scores, and that the tendancy for scores to decrease over time in the program is accurate.

## Summary Report

What did the data reveal?
- The majority of patients are taking a low number of tests (average ~3.5). This meant we needed to find a way to measure progress that did not rely on number of tests. Reference chart [I].
- The individual scores did not paint a pattern that we could analyze as an indication of progress. Reference chart [II]. There were either not enough data points per patient or the plotting of scores was not a sufficient indicator of progress.
- To analyze progress, we created a new metric that retrieved the amount of time that a patient was receiving treatment. The length of time was found by the measuring the days between the creation date of the account and the most recent test taken. This eliminated the problem of inconsistent testing dates between patients and differing numbers of tests and further allowed us to compare patients who had been in the program for the same amounts of time.
- We then aggregated the patients by duration of time in the program, and derived insights according to two measures:
    - [1] The change between the first test taken and the most recent test
    - [2] The means, minima, and maxima of each each patient's scores
- The first metric showed us that patients with large increases in their scores between their first and most recent tests tended to have larger increseases the longer they stayed in the program. The patients with the largest decreases in scores from the first test to the most recent test conversely tended to have larger decreases the longer they stayed in the program. On average, the tendancy was that patient scores were decreasing slightly between the first and last test. On initial review, this told us that patients were gaining momentum in one direction 
    - Advantages: We are comparing only two tests for every patient which provides consistency between patients with differing numbers of tests.
    - Disadvantages: This metric ignores tests taken inbetween which may have large fluctuations. Looking at the destribution of standard deviaitions in chart [VII], however, this does not affect the majority of patients.
- To handle some of the issues in the first approach, I decided to aggregate each patient by their average score before further aggregating the average for each duration of time as before. This metric showed that the average, minima, and maxima of patient scores aggregated by duration of time are all improving as duration increases. The minima had a much higher correlation than the maxima, which would indicate that patients with generally lower scores tend to improve more easily over time.
    - Advantages: This takes into account all of the tests that each patient has taken, using the mean instead of median to weight outliers in the metric.
    - Disadvantages: Rather than looking at the improvement of each individual and applying that to a larger scale, we are taking the average of each score for duration of time. This means we are viewing the correlation between days in the program and and average score rather than an improvement of score.

Conclusion: We can reasonably assume based on these insights that as patients work through exercises on the platform over a longer period of time, their scores on the GAD-7 are improving. This means that patients' mental health are generally improving through use of the platform. The longest period of time that a patient was recorded between the first and most recent test was 180 days, so it will be interesting to see how these insights change with more data and time.
